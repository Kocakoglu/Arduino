
## 2. Veriyi ön işleme (Preprocessing) adımlarını açıklayarak uygulayınız. (20 Puan)

# - Eksik veriler (missing values) için uyguladığınız stratejileri açıklayınız.

Veri kümesi üzerinde eksik veri (missing value) analizi yapılmıştır. pandas kütüphanesinin isnull().sum() fonksiyonu kullanılarak her sütundaki eksik değerlerin sayısı ve oranı hesaplanmıştır. Sonuçlara göre:

Veri setinde hiçbir sütunda eksik veri bulunmamaktadır.
Bu nedenle herhangi bir doldurma (imputation) stratejisi uygulamaya gerek kalmamıştır.

Alternatif olarak eksik veri olması durumunda şu stratejiler düşünülebilirdi:

Sayısal sütunlarda ortalama, medyan veya mod ile doldurma

Kategorik sütunlarda mod ile doldurma

Gelişmiş tekniklerde KNN veya regresyon tabanlı doldurma

Veya eksik satır/sütunların silinmesi (dropna())

### - Sayısal verilerde standartlaştırma (Standardization) veya normalizasyon (Normalization) işlemleri yaptıysanız belirtiniz.

Veri setindeki sayısal değişkenlerin farklı ölçeklerde olması (örneğin, GRE puanı 340’a kadar çıkarken CGPA en fazla 10’dur) modelleme sürecinde dengesizlik yaratabileceğinden, tüm sayısal değişkenlere standardizasyon (z-score normalization) uygulanmıştır.

Kullanılan yöntem: StandardScaler()
Amaç: Her sütunun ortalamasını 0, standart sapmasını 1 yapmak

Bu projede kullanılan veri kümesi yalnızca sayısal (tabular) verilerden oluşmaktadır. Her satır bir öğrencinin başvurusunu temsil ederken, sütunlar sayısal ya da ikili (binary) niteliklere sahiptir.

Bu nedenle projede görsel veri (görüntü, resim, piksel matrisi) bulunmamaktadır.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 1. Gereksiz sütunun kaldırılması
df = df.drop("Serial No.", axis=1, errors='ignore')

# 2. Eksik veri kontrolü ve tablo oluşturma
missing_df = pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Percentage': (df.isnull().sum() / len(df)) * 100
})
print("🔍 Eksik Veri Tablosu:")
display(missing_df)

# 3. Sayısal özet (describe) tablosu
print("\n📊 Sayısal Özelliklerin Temel İstatistikleri:")
display(df.describe().T)

# 4. Kategorik değişkenin (Research) dağılımı
research_counts = df['Research'].value_counts().rename(index={0: "No Research", 1: "Has Research"})
plt.figure(figsize=(6,4))
sns.barplot(x=research_counts.index, y=research_counts.values, palette="Set2")
plt.title("Araştırma Deneyimi Dağılımı")
plt.ylabel("Öğrenci Sayısı")
plt.xlabel("Araştırma Durumu")
plt.tight_layout()
plt.show()

# 5. Hedef ve özelliklerin ayrılması
X = df.drop("Chance of Admit ", axis=1)
y = df["Chance of Admit "]

# 6. Standardizasyon
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# 7. Standardize edilmiş verinin histogramları
X_scaled_df.hist(bins=20, figsize=(15, 10))
plt.suptitle("Standardize Edilmiş Özelliklerin Dağılımı", fontsize=16)
plt.tight_layout()
plt.show()

# 8. Boxplot ile görselleştirme
plt.figure(figsize=(12, 6))
sns.boxplot(data=X_scaled_df)
plt.title("Standardize Edilmiş Özelliklerin Boxplot Grafiği")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#  9. İlk 5 satır örnek
print("\n📐 Standardize Edilmiş Özellikler (İlk 5 Satır):")
display(X_scaled_df.head())
```

```python
import os
import pandas as pd

# processed_df'yi oluştur
processed_df = pd.concat([X_scaled_df, y.reset_index(drop=True)], axis=1)

# Klasör oluştur ve dosya kaydet
dir_path = "/content/drive/MyDrive/YZ"
if not os.path.exists(dir_path):
    os.makedirs(dir_path)

processed_path = os.path.join(dir_path, "processed_admission_data.csv")
processed_df.to_csv(processed_path, index=False)

# İlk satırları göster
processed_df.head()
```

## 3. Modelleme ve Özellik Mühendisliği Adımlarını Açıklayarak En Az İki Farklı Model Uygulayınız. (40 Puan)

- Kullanmayı tercih ettiğiniz modelleri belirtiniz. Bunlar makine öğrenmesi (k-Nearest Neighbors - kNN, Support Vector Machine - SVM, Decision Tree, Multi-Layer Perceptron - MLP) veya derin öğrenme (Convolutional Neural Network - CNN, Long Short-Term Memory - LSTM, transfer öğrenme) tabanlı olabilir.
- Özellik çıkarımı (feature extraction) veya özellik seçimi (feature selection) uyguladıysanız hangi yöntemleri kullandığınızı açıklayınız (örneğin: histogram, PCA - Principal Component Analysis, FFT - Fast Fourier Transform, GLCM - Gray Level Co-occurrence Matrix).
- Uygulanan yöntemlerin veri setinize neden uygun olduğunu belirtiniz.
- Modelleri eğitiniz ve değerlendiriniz. Kullanılan başarı metriklerini (accuracy, precision, recall, F1-score, confusion matrix, ROC-AUC vb.) belirtiniz.
- Hiperparametre optimizasyonu yaptıysanız (örneğin: GridSearchCV) açıklayınız.
- En az iki modeli performans açısından karşılaştırınız.


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42)

# 2. Model 1: Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_preds = lr_model.predict(X_test)

# 3. Model 2: k-Nearest Neighbors (kNN) Regression
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train, y_train)
knn_preds = knn_model.predict(X_test)

# 4. Performans metrikleri
def evaluate_model(name, y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"{name} Model")
    print(f"  MSE: {mse:.4f}")
    print(f"  R² : {r2:.4f}\n")
    return mse, r2

lr_mse, lr_r2 = evaluate_model("Linear Regression", y_test, lr_preds)
knn_mse, knn_r2 = evaluate_model("kNN Regression", y_test, knn_preds)

# 5. Gerçek vs Tahmin Grafikleri
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=y_test, y=lr_preds, alpha=0.7)
plt.xlabel("Gerçek Değerler")
plt.ylabel("Tahmin Değerleri")
plt.title("Linear Regression - Gerçek vs Tahmin")

plt.subplot(1, 2, 2)
sns.scatterplot(x=y_test, y=knn_preds, color="orange", alpha=0.7)
plt.xlabel("Gerçek Değerler")
plt.ylabel("Tahmin Değerleri")
plt.title("kNN Regression - Gerçek vs Tahmin")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Regresyon Modelleri
from sklearn.linear_model import LinearRegression, Lasso, Ridge, BayesianRidge, ElasticNet, HuberRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.linear_model import SGDRegressor
from xgboost import XGBRegressor

# Özellik ve hedef değişken ayrımı

# Özellik ve hedef değişken ayrımı
X = df.drop(['Chance of Admit '], axis=1)
y = df['Chance of Admit ']

# Standardizasyon
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Eğitim/test ayrımı
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42, shuffle=True)

# Modeller listesi (CatBoost ve LightGBM çıkarıldı)
models = [
    ['Linear Regression', LinearRegression()],
    ['Ridge', Ridge()],
    ['Lasso', Lasso()],
    ['BayesianRidge', BayesianRidge()],
    ['ElasticNet', ElasticNet()],
    ['HuberRegressor', HuberRegressor()],
    ['Decision Tree', DecisionTreeRegressor()],
    ['Random Forest', RandomForestRegressor()],
    ['AdaBoost', AdaBoostRegressor()],
    ['GradientBoosting', GradientBoostingRegressor()],
    ['KNeighbors', KNeighborsRegressor(n_neighbors=5)],
    ['SVR', SVR()],
    ['SGDRegressor', SGDRegressor()],
    ['XGBoost', XGBRegressor(verbosity=0)]
]

# Model değerlendirme fonksiyonu
def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    r2 = r2_score(y_test, preds)
    return [name, rmse, r2]

# Tüm modelleri değerlendir
results = []
for name, model in models:
    result = evaluate_model(name, model, X_train, y_train, X_test, y_test)
    results.append(result)

# Sonuçları DataFrame olarak derle ve sırala
results_df = pd.DataFrame(results, columns=["Model", "RMSE", "R2"])
results_df = results_df.sort_values(by="RMSE", ascending=True).reset_index(drop=True)

# Sonuçları yazdır
print("Model Performans Karşılaştırması:")
print(results_df)

# Görselleştirme: RMSE Karşılaştırması
plt.figure(figsize=(12, 6))
sns.barplot(x="RMSE", y="Model", data=results_df, palette="viridis")
plt.title("Regresyon Modelleri - RMSE Karşılaştırması")
plt.xlabel("Root Mean Squared Error (RMSE)")
plt.ylabel("Model")
plt.tight_layout()
plt.show()

# Özellik önem sıralaması (Random Forest ile)
rf = RandomForestRegressor()
rf.fit(X_scaled_df, y)
importance_frame = pd.DataFrame()
importance_frame['Features'] = X.columns
importance_frame['Importance'] = rf.feature_importances_
importance_frame = importance_frame.sort_values(by='Importance', ascending=True)

# Görselleştirme: Özellik önemi
plt.figure(figsize=(10, 6))
sns.barplot(x="Importance", y="Features", data=importance_frame, palette="mako")
plt.title("Özelliklerin Göreli Önemi (Random Forest ile)")
plt.xlabel("Önemi")
plt.ylabel("Özellik")
plt.tight_layout()
plt.show()


```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, r2_score

# Eğitim/test veri seti zaten tanımlı (X_train, X_test, y_train, y_test)

# Derin öğrenme modeli (MLP)
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='linear')  # regresyon çıktısı
])

# Derleme
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# Eğitim
history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_split=0.1, verbose=0)

# Tahmin
y_pred_dl = model.predict(X_test).flatten()

# Değerlendirme
mse_dl = mean_squared_error(y_test, y_pred_dl)
r2_dl = r2_score(y_test, y_pred_dl)

print(f"Derin Öğrenme Modeli Sonuçları:")
print(f"  MSE: {mse_dl:.4f}")
print(f"  R² : {r2_dl:.4f}")
```
